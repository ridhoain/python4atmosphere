# Gabungkan data .nc menjadi 1 file saja

```python
import os
import xarray as xr
from netCDF4 import Dataset

def combine_nc_files(folder_path, output_path):
    try:
        # Mengecek apakah folder_path adalah direktori yang valid
        if os.path.isdir(folder_path):
            # Mendapatkan list dari nama-nama file dalam folder
            files = os.listdir(folder_path)
            # Memfilter hanya file dengan ekstensi .nc4
            nc_files = [file for file in files if file.endswith('.nc4')]
            
            if len(nc_files) == 0:
                print("Tidak ada file .nc4 dalam folder.")
                return
            
            # Membaca file pertama untuk mendapatkan struktur
            first_nc = xr.open_dataset(os.path.join(folder_path, nc_files[0]))
            combined_data = first_nc
            
            # Looping melalui setiap file .nc4
            for nc_file in nc_files[1:]:
                print("Reading:", nc_file)
                nc_data = xr.open_dataset(os.path.join(folder_path, nc_file))
                combined_data = xr.concat([combined_data, nc_data], dim='time')
                nc_data.close()
            
            # Menyimpan data gabungan ke dalam file .nc
            combined_data.to_netcdf(output_path)
            print("File gabungan berhasil disimpan:", output_path)
            
            # Menutup file pertama yang telah dibuka
            first_nc.close()
        else:
            print(folder_path, "bukan merupakan direktori yang valid.")
    except Exception as e:
        print("Error:", e)

# Path ke direktori yang berisi file .nc4 yang ingin Anda gabungkan
folder_path = r'C:\Users\ck818273\Downloads\downthemall'
# Path untuk menyimpan file gabungan
output_path = r'C:\Users\ck818273\Music\your_download_folder_path_here\latihanzzz.nc'

combine_nc_files(folder_path, output_path)
```

# Cek apakah data telah berhasil di gabung dan sesuai

```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
# Alamat file dan nama file
data = xr.open_dataset('C:\\Users\\ck818273\\Music\\your_download_folder_path_here\\latihanzzz.nc')
```

```python
latitude = data['lat'][:]
longitude = data['lon'][:]
pre = data['precipitation'][:]  # Make sure to extract the data from the HDF5 dataset

tes = pre[0, :, :]
```

```python
import cartopy.feature as cfeature
import h5py
import numpy as np
import xarray as xr
from pathlib import Path
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as creature

# Create a Cartopy projection
projection = ccrs.PlateCarree()

# Create a figure and axis with the specified projection
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw={'projection': projection})

# Plot data using the 'pcolormesh' method, which is better suited for gridded data
lon2d, lat2d = np.meshgrid(longitude, latitude)
img = ax.pcolormesh(lon2d, lat2d, tes.T, transform=projection, cmap='Blues')  # Transpose tes here

# Add colorbar
plt.colorbar(img, ax=ax, label='Precipitation (mm/hr)', shrink=0.5, pad=0.05)

# Add features to the map
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.LAND, facecolor='lightgrey')

# Set the extent of the plot to the min and max of your longitude and latitude
ax.set_extent([longitude.min(), longitude.max(), latitude.min(), latitude.max()], crs=projection)

# Add gridlines
gl = ax.gridlines(crs=projection, draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False

ax.set_title('Precipitation Data', size=16)

# Save the figure as JPEG
output_file_path = 'precipitation_latihan1.jpeg'
plt.savefig(output_file_path)  # Remove the quality parameter
plt.show()
```
